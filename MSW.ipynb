```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy.special import factorial as scipy_factorial
import time
import pandas as pd
import random
from scipy.integrate import quad
```

```python
# Faktoriál čisel

def factorial_py(n):
    if n == 0 or n == 1:
        return 1
    else:
        result = 1
        for i in range(2, n + 1):
            result *= i
        return result

def factorial_numpy(n):
    if n == 0 or n == 1:
        return 1
    else:
        return np.prod(np.arange(2, n + 1))

def factorial_scipy(n):
    return scipy_factorial(n, exact=True)


n = 50

# Standarni Python
started = time.time()
result_py = factorial_py(n)
end_py = time.time() - started

# Numpy
started = time.time()
result_np = factorial_numpy(n)
end_np = time.time() - started

# Scipy
started = time.time()
result_scipy = factorial_scipy(n)
end_scipy = time.time() - started

print(f"Faktoriál čisla {n} je {result_scipy}")
print(f"Výpočet Python trvá {end_py}")
print(f"Výpočet Numpy trvá {end_np}")
print(f"Výpočet Scipy trvá {end_scipy}")


fast = min(end_py, end_np, end_scipy)
if fast == end_py:
  print(f"Nejrychlejší je Standardní Python")
elif fast == end_np:
  print("Nejrychlejší je Numpy")
else:
  print("Nejrychlejší je Scipy")
```

```python
# Integrál funkce Riemannův součet

def funkce(x):
    return x ** 2

def integrate_py(a, b, n):
    dx = (b - a) / n
    integral = 0
    for i in range(n):
        x = a + i * dx
        integral += funkce(x) * dx
    return integral

a = 0
b = 1
n = 1000

start = time.time()
result_py = integrate_py(a, b, n)
end_py = time.time() - start

x_val = np.linspace(a,b,n)
y_val = funkce(x_val)

start = time.time()
result_np = np.trapz(y_val, x_val)
end_np = time.time() - start

print(f"Hodnota integrálu funkce f(x) na intervalu [{a}, {b}] je\nStandarni Python: {result_py}\nNumpy: {result_np}")
print(f"Výpočet pro Python trvá {end_py:.7f} sekund")
print(f"Výpočet pro Numpy trvá {end_np:.7f} sekund")

fast = min(end_py, end_np)
if fast == end_py:
  print(f"Nejrychlejší je Standardní Python")
else:
  print("Nejrychlejší je Numpy")
```

```python
# Kvadratická funkce

def kvad_funkce(a, b, c):
    discriminant = b**2 - 4*a*c
    if discriminant > 0:
        x1 = (-b + discriminant ** 0.5) / (2*a)
        x2 = (-b - discriminant ** 0.5) / (2*a)
        return x1, x2
    elif discriminant == 0:
        x = -b / (2*a)
        return x,
    else:
        return None

def kvad_np(a,b,c):
  koef = [a,b,c]
  solution = np.roots(koef)
  return solutions

a = 1
b = -3
c = 2
start = time.time()
solutions = kvad_funkce(a, b, c)
end_py = time.time() - start

start = time.time()
reseni_np = kvad_np(a,b,c)
end_np = time.time() - start

if solutions:
    print("Výpočet kvadratické funkce je Standarni Python", solutions)
else:
    print("Funkce nemá řešení.")

print("Výpočet kvadratické funkce Numpy je ", reseni_np)
print(f"Výpočet pro Python trvá {end_py:.7f}")
print(f"Výpočet pro Numpy trvá {end_np:.7f}")

fast = min(end_py, end_np)
if fast == end_py:
  print(f"Nejrychlejší je Standardní Python")
else:
  print("Nejrychlejší je Numpy")
```

```python
# Nasobení matric

def matrix_multiply(matrix1, matrix2):
    if len(matrix1[0]) != len(matrix2):
        raise ValueError("Počet sloupců je jiný, zkuste to znovu")

    result = [[0 for i in range(len(matrix2[0]))] for _i in range(len(matrix1))]

    for i in range(len(matrix1)):
        for j in range(len(matrix2[0])):
            for k in range(len(matrix2)):
                result[i][j] += matrix1[i][k] * matrix2[k][j]

    return result

def matrix_np(mat1, mat2):
  result = np.dot(mat1, mat2)
  return result


matrix1 = [[1, 2],
           [4, 5]]
matrix2 = [[7, 8],
           [10, 11]]

# Python
start1 = time.time()
result_py = matrix_multiply(matrix1, matrix2)
end_py = time.time() - start1

# Numpy
start2 = time.time()
result_np = matrix_np(matrix1, matrix2)
end_np = time.time() - start2

print("Výsledek nasobení matric pres Standarni Python:")
for row in result_py:
    print(row)
print(f'Výsledek nasobení matric pres Numpy:\n{result_np}')

print(f"Výpočet pro Python trvá {end_py:.7f} sekund")
print(f"Výpočet pro Numpy trvá {end_np:.7f} sekund")

fast = min(end_py, end_np)
if fast == end_py:
  print(f"Nejrychlejší je Standardní Python")
else:
  print("Nejrychlejší je Numpy")
```

```python
#Linearni rovnice
def linearni_rovnice(a, b):
    if a == 0:
        if b == 0:
            return "Rovnice má hodně řešení"
        else:
            return "Rovnice nemá řešení"
    else:
        x = -b / a
        return x

def linearni_np(a, b):
    if a == 0:
        if b == 0:
            return "Rovnice má hodně řešení."
        else:
            return "Rovnice nemá řešení."
    else:
        x = np.linalg.solve([[a]], [[-b]])
        return x[0]

a = 6
b = 12

# Python
start = time.time()
result = linearni_rovnice(a, b)
end_py = time.time() - start

# Numpy
start = time.time()
result_np = linearni_np(a, b)
end_np = time.time() - start

print(f"Výsledek lineární rovnice f(x)= {a}x + {b}\nje {result} v Standard Python")
print(f'a v Numpy je {result_np}')
print(f"Výpočet trvá pro Python {end_py:.7f} sekund")
print(f"Výpočet trvá pro Numpy {end_np:.7f} sekund")

fast = min(end_py, end_np)
if fast == end_py:
  print(f"Nejrychlejší je Standardní Python")
else:
  print("Nejrychlejší je Numpy")
```

```python
def function(x):
    return x**2

def numerical_derivative_adaptive(func, x, h=1e-5, tol=1e-6):

    dx = h
    prev_derivative = (func(x + dx) - func(x - dx)) / (2 * dx)
    while True:
        dx /= 2
        derivative = (func(x + dx) - func(x - dx)) / (2 * dx)
        if np.abs(derivative - prev_derivative) < tol:
            return derivative
        prev_derivative = derivative


def analytical_derivative(x):
    return 2 * x


x = 18

adaptive_derivative = numerical_derivative_adaptive(function, x)


analytical_derivative_result = analytical_derivative(x)

print("Numerická derivace s adaptabilním krokem:", adaptive_derivative)
print("Analytická derivace:", analytical_derivative_result)
```

## **Uloh 7**

```python
def estimate_pi(num_samples):
    count = 0
    for _ in range(num_samples):
        x = random.uniform(-1, 1)
        y = random.uniform(-1, 1)
        if x**2 + y**2 <= 1:
            count += 1
    return (count / num_samples) * 4

num_samples = 1000000

pi_estimate = estimate_pi(num_samples)
print("Odhad hodnoty π pomocí Monte Carlo simulace:", pi_estimate)
```
